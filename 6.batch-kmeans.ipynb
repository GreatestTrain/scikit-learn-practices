# MiniBatchKMeans

```python
import pandas as pd
import numpy as np
from sklearn.cluster import MiniBatchKMeans
```

Es conveniente utilizar `MiniBatchKMeans` si contamos con una decente cantidad de poder de c√≥mputo pero no con mucha memoria RAM.

```python
dataset = pd.read_csv('data/candy.csv')
dataset.head(5)
```

```python
X = dataset.drop(columns=['competitorname'])
```

```python
import os
```

```python
os.environ['OMP_NUM_THREADS'] = "1"
```

```python
kmeans = MiniBatchKMeans(n_clusters=4)
kmeans.fit(X)
pred = kmeans.predict(X)
print("Prediction \n" + "-"*16)
print(pred)
print(np.unique(kmeans.labels_))
```

```python
X['group'] = pred
X.head()
```
